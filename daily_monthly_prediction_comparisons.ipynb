```python
# Load dependencies
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
parameter_list = [
    'Type',
    'SVM',
    'C',
    'Gamma',
    'Epsilon',
    'Score',
    'MSE'
]
dpA_params = [
    'Daily',
    'RBF',
    10,
    0.1,
    0.1,
    0.2878,
    0.00052
]
dpB_params = [
    'Daily',
    'RBF',
    10,
    0.1,
    0.1,
    0.7639,
    0.0608
]
dpC_params = [
    'Daily',
    'RBF',
    5,
    0.1,
    0.1,
    0.3146,
    0.00000013
]
dpD_params = [
    'Daily',
    'RBF',
    10,
    0.1,
    0.1,
    0.4970,
    0.0000098
]
mpA_params = [
    'Monthly',
    'RBF',
    5,
    0.1,
    0.1,
    0.7528,
    0.00026
]
mpB_params = [
    'Monthly',
    'RBF',
    1,
    0.1,
    0.01,
    0.7774,
    0.0358
]
mpC_params = [
    'Monthly',
    'NONE',
    'NONE',
    'NONE',
    'NONE',
    'NONE',
    'NONE'
]
mpD_params = [
    'Monthly',
    'RBF',
    10,
    0.1,
    0.001,
    0.6434,
    0.0000051
]
```

```python
regression_dict = {
    'pA_D': dpA_params,
    'pB_D': dpB_params,
    'pC_D': dpC_params,
    'pD_D': dpD_params,
    'pA_M': mpA_params,
    'pB_M': mpB_params,
    'pC_M': mpC_params,
    'pD_M': mpD_params
}
```

```python
regressions = pd.DataFrame(regression_dict,index=parameter_list)
regressions
```

Based on the table above, monthly regressions have significantly better scores and lower MSE values than their daily counterparts. This, combined with the complications of running projections on a daily, rather than monthly, basis is why the remainder of the project will only focus on the monthly regressions.

Note that there is no pC monthly model; there were not enough data points to justify a model as that contaminant did not start getting checked for until mid 2017. Having accurate models for pA, pB, and pD is more than was originally expected
